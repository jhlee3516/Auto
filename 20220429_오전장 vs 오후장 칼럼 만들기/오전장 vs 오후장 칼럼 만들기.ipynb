### 2022/03/29 → 2022-03-29

```python
# 방법1
df = pd.read_csv('data.csv')
df['주문일자'] = df['주문일자'].str.replace('/','-')
print(df.loc[0, '주문일자'])
print(type(df.loc[0, '주문일자']))
```

```python
# 방법2
df = pd.read_csv('data.csv')
df['주문일자'] = pd.to_datetime(df['주문일자'])
print(df.loc[0, '주문일자'])
print(type(df.loc[0, '주문일자']))
```

# ② 오전장 vs 오후장
오전장: 09:00~12:00  
오후장: 12:00~13:30

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
df['매수시간'] = pd.to_datetime(df['매수시간'])
df.loc[0, '매수시간'].time()
```

```python
datetime.time(12, 0, 0)
```

```python
df.loc[0, '매수시간'].time() < datetime.time(12, 0, 0)
```

```python
df['매수시간'] = df['매수시간'].apply(lambda x : x.time())
df.head()
```

```python
df[df['매수시간'] > datetime.time(12, 0, 0)]
```

```python
df['오전장'] = df['매수시간'] < datetime.time(12, 0, 0)
df.head()
```
