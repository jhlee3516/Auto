```python
import FinanceDataReader as fdr
from datetime import timedelta
```

```python
stock_code = '028300'
df_finance = fdr.DataReader(symbol=stock_code)
df_finance.tail()
```

```python
df_ceiling = df_finance[df_finance['Change'] > 0.29]
df_ceiling
```

```python
NextClose = []
for index in df_ceiling.index:
    date = index + timedelta(days=1)
    next_close = df_finance.loc[date,'Change']
    next_close = round(next_close*100,2)
    NextClose.append(next_close)
df_ceiling['NextClose'] = NextClose
df_ceiling
```
