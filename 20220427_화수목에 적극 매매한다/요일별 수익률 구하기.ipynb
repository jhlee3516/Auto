# ② Methods

### Import

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Load

```python
df = pd.read_csv('data.csv')
df.head()
```

### string → datetime

```python
df.loc[0, '매수시간']
```

```python
type(df.loc[0, '매수시간'])
```

```python
df['매수시간'] = pd.to_datetime(df['매수시간'])
df.head()
```

```python
type(df.loc[0, '매수시간'])
```

```python
df.loc[0, '매수시간'].day_name()
```

### 요일 칼럼 만들기

```python
df['주문일자'] = pd.to_datetime(df['주문일자'])
df.head()
```

```python
# 방법 1
df['요일'] = df['주문일자'].apply(lambda x : x.day_name())
df.head()
```

```python
# 방법 2
df['요일2'] = df['주문일자'].dt.day_name()
df.head()
```

### 요일별 수익률 구하기

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df.groupby('요일')['수익률'].mean().plot.bar()
plt.axhline(y=0, color='r', linewidth=3)
plt.title('Mean', size=15)
plt.show()
```

```python
df.groupby('요일')['수익률'].median().plot.bar()
plt.axhline(y=0, color='r', linewidth=3)
plt.title('Median', size=15)
plt.show()
```
