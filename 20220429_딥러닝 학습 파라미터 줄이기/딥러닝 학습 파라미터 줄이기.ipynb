# ① Introduction
  과대적합은 딥러닝 모델이 훈련 데이터에만 특화된 패턴을 학습하기 시작했다는 의미입니다.  
  훈련 데이터에만 특화된 패턴은 새로운 데이터와 관련성이 적어 잘못된 판단을 하게 만듭니다.  
  과대적합을 해결하려면 더 많은 훈련 데이터를 모아야 합니다.  
  하지만 현실적으로 더 많은 훈련 데이터를 모으는 것이 불가능할 때 차선책은 모델이 저장할 수 있는 정보에 제약을 가하는 것입니다.   
  즉, 모델의 학습 파라미터 수를 줄이는 것입니다.   
  유튜브: https://youtu.be/zPJNdAG6T3s

# ② Methods

### Import

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adagrad

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

import pandas as pd
import numpy as np
```

### 샘플 개수
훈련 데이터의 샘플 개수는 141개로 작기 때문에 과대적합될 가능성이 높습니다.  
따라서 모델의 학습 파라미터 수를 줄여줍니다.

```python
df = pd.read_csv('data.csv')
len(df)
```

```python
df['수익률'] = df['수익률'].apply(lambda x : 1 if x > 0 else 0 )
df['수익률'].unique()
```

```python
X = df[['뉴스기사', '직전 거래대금', '최대거래대금', '양봉개수']]
y = df['수익률']

X = X.to_numpy()
y = y.to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

model = Sequential()
model.add(Dense(4, input_dim=4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
optimizer = Adagrad(lr=0.2)

model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=40)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```

# ③ Results
Layer가 12, 8, 1 일 때 Score는 55%였는데
